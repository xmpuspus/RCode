# Season Stats

```r
data = read.csv('data/seasons_stats.csv')
```

```r
head(data)
```

```r
summary(data)
```

```r
names(data)
```

### Player with the Highest 3-Point Attempt Rate

```r
data[c("Player", 'X3PAr')][which(data$X3PAr == max(data$X3PAr, na.rm = TRUE)),]
```

```r
head(data[c("Player", 'X3PAr')][order(data$X3PAr, decreasing = TRUE),])
```

### Player with the Highest Free Throw Rate

```r
data[c("Player", 'FTr')][which(data$FTr == max(data$FTr, na.rm = TRUE)),]
```

```r
head(data[c("Player", 'FTr')][order(data$FTr, decreasing = TRUE),], 10)
```

### Players Who Scored the Highest Each Year

```r
# Who are the players with the highest 
df_agg_teams = do.call(rbind, lapply(split(data, as.factor(data$Player)), function(x) {return(x[which.max(x$PTS),])}))
df_agg_teams
```

```r
unique(data$Player)[order(data$Player)]
```

### Year LeBron James Scored the Highest

```r
df_agg_teams[which(df_agg_teams$Player == 'LeBron James'),]
```

### Year Michael Jordan Scored the Highest

```r
df_agg_teams[which(df_agg_teams$Player == 'Michael Jordan*'),]
```

```r
# Player efficiency rating of Kobe Bryant in the year where his MP is the lowest? 
kb_data = data[which(data$Player == 'Kobe Bryant'),]
kb_data[which.min(kb_data$MP),]$PER
```
